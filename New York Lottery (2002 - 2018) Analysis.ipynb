```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('/Users/asamarakone/Desktop/Data_Sources/Lottery_Mega_Millions_Winning_Numbers__Beginning_2002.csv')
```

```python
df.drop('Multiplier',axis = 1, inplace = True)
df.head()
```

```python
firstNumbers = []
secondNumbers = []
thirdNumbers = []
fourthNumbers = []
fifthNumbers = []
sumofwinningNumbers =[]

for winning_numbers_string in df['Winning Numbers']:
    winning_numbers_list = winning_numbers_string.split()
    firstNumbers.append(int(winning_numbers_list[0]))
    secondNumbers.append(int(winning_numbers_list[1]))
    thirdNumbers.append(int(winning_numbers_list[2]))
    fourthNumbers.append(int(winning_numbers_list[3]))
    fifthNumbers.append(int(winning_numbers_list[4]))
    sumofwinningNumbers.append(sum([int(x) for x in winning_numbers_list]))
    
df['1st winning number'] = firstNumbers
df['2nd winning number'] = secondNumbers
df['3rd winning number'] = thirdNumbers
df['4th winning number'] = fourthNumbers
df['5th winning number'] = fifthNumbers
df['Sum of 5 winning numbers'] = sumofwinningNumbers

months = []
days = []
years = []

for date in df['Draw Date']:
    months.append(int(date.split('/')[0]))
    days.append(int(date.split('/')[1]))
    years.append(int('20' + date.split('/')[2]))
    
df['Draw Month'] = months
df['Draw Day'] = days
df['Draw Year'] = years
```

```python
def plotWinningNumberPieChart(placeholder):
    
    even_count = 0
    prime_count = 0
    not_prime_count = 0

    for num in df['{} winning number'.format(placeholder)]:
        if (num%2) == 0:
            even_count = even_count + 1
        else:
            if (num > 0) & (len(factorize(num)) == 2):
                prime_count = prime_count + 1
                
            else:
                not_prime_count = not_prime_count + 1
        
    plt.pie([even_count,not_prime_count,prime_count], labels = ['Even','Not Prime','Prime'],startangle = 0)
    plt.axis('equal')
    plt.title('Percentage of {} winning numbers that are even, prime, or not prime'.format(placeholder))
    
def factorize(n):

    if n < 0:
        somepossibleFactors = [x for x in range(2,-n+1)]
        factors = [-1]
    
    elif n > 0:
        somepossibleFactors = [x for x in range(2,n+1)]
        factors = [1]
        
    else:
        somepossibleFactors = []
        factors = [0,1]
        
    for possibleFactor in somepossibleFactors:
        while (n%possibleFactor == 0):
            factors.append(possibleFactor)
            n = n/possibleFactor

    return(factors)
```
