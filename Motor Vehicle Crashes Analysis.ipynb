```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('/Users/asamarakone/Desktop/Data_Sources/Motor_Vehicle_Crashes_-_Vehicle_Information__Three_Year_Window.csv')
```

```python
s = df['Year'].value_counts()
plt.bar(s.index,s.values)
s
```

```python
s = df['Vehicle Body Type'].value_counts()[0:10]
plt.bar(s.index,s.values)
plt.xticks(rotation = 'vertical')
print(s)
```

```python
s = df['Vehicle Year'].value_counts()
s = s.sort_index(ascending = False)
plt.bar(s.index,s.values)
plt.xticks(rotation = 'vertical')


print('[Vehicle Year] | [Frequency]')

for i in range(0,25):
    print('    ',s.index[i],'   |    ', s.values[i])
```
